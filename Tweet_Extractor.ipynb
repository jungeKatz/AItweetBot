#### The function that extracts the tweets 

```python
def get_tweets(username):
  auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
  auth.set_access_token(access_key, access_secret)
  api = tweepy.API(auth)
  
  all_tweets = []

  tweets = api.user_timeline(screen_name=username ,
                             count=200,
                           include_rts = False,
                            tweet_mode='extended')
  all_tweets.extend(tweets)
  oldest_id = tweets[-1].id
  while True:
    tweets = api.user_timeline(screen_name=username, 
                           # 200 is the maximum allowed count
                           count=200,
                           include_rts = False,
                           max_id = oldest_id - 1,
                           # Necessary to keep full_text 
                           # otherwise only the first 140 words are extracted
                           tweet_mode = 'extended'
                           )
    if len(tweets) == 0:
        break
    oldest_id = tweets[-1].id
    all_tweets.extend(tweets)
    

  #print(tweets)
  tmp=[]
  tweets_for_csv = [tweet.full_text for tweet in all_tweets] 
  for j in tweets_for_csv:
    tmp.append(j)
  # print(tmp)
  return tmp
    
  
	
	
```

#### Calling the function and adding tweets to an array

```python
def remove_urls(vTEXT):
    vTEXT = re.sub(r'(https|http)?:\/\/(\w|\.|\/|\?|\=|\&|\%)*\b', '', vTEXT, flags=re.MULTILINE)
  
    return(vTEXT)
```

```python
def remove_names(tweet):
  tweet = tweet.split("-")[0]
  return tweet
```

```python
words = set(nltk.corpus.words.words())
```

```python
if __name__ == '__main__':

	# Here goes the twitter handle for the user
	# whose tweets are to be extracted.
  combined_array =[]
  # more users can be added to this array and their tweets can be saved
  users = ['SadhguruJV','oshoonline','radhanathswami','OshoQuotes','thesecret','LAWOFATTRACTlON',
           'TheQuoteBoook_','SeffSaid','_SuccessMinded_','billionair_key','billionaire_key','billionaire_sec'
           ,'Motiv8Central','SucessCentral','dailymotiv8ion','M2Succeed','Sahi_Soch',
           'Cow__Momma','DilliDurAst'] #  for example ["user1" , "user2"....] and so on 
  for user in users:
   tweet_array=get_tweets(user)
   for tweet in tweet_array:
    
    tweet = re.sub("@[A-Za-z0-9]+","",tweet)
    tweet = re.sub("#[A-Za-z0-9]+","",tweet)
    tweet = re.sub("sadhg[A-Za-z0-9]+","",tweet)
    tweet = re.sub("osho[A-Za-z0-9]+","",tweet)
    tweet = remove_names(tweet)
    tweet = remove_urls(tweet)
    
    tweet = " ".join(w for w in nltk.wordpunct_tokenize(tweet) if w.lower() in words or not w.isalpha())
    print(tweet)
    combined_array.append(tweet)
   
```

```python
header =["tweet"]
data = combined_array
```
